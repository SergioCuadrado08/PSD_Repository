```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
A = 10
frec = 10

t_1 = np.linspace(0, 1, 512)
t_2 = np.linspace(0, 1, 512)
func_A = A * np.sin(2*np.pi * frec*t_1 ) + 1
func_B = (A * np.sin(2*np.pi * 233*t_2 )) + func_A
```

```python
plt.plot(t_1,func_A)
plt.plot(t_2,func_B)
```

```python
fft_A = np.fft.fft(func_A)
fft_B = np.fft.fft(func_B)

frecfftA = np.fft.fftfreq(512)
frecfftB = np.fft.fftfreq(len(fft_B),2048)

```

```python
plt.plot ((1/frecfftB),np.abs(fft_B))
```
