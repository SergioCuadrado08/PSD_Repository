```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interactive
from ipywidgets import interactive_output, VBox

def plot_signals_and_convolution(f2):
# Definir el tiempo de muestreo
  fs = 5000  # Frecuencia de muestreo
  t = np.linspace(0, 1, fs)  # 1 segundo de duración
  tc = np.linspace(0, 1, fs*2-1)
  # Definir señales
  sig1 = 10 * np.sin(2 * np.pi * 5 * t) + 10 * np.sin(2 * np.pi * f2 * t)  # Señal combinada
  sig2 = 10 * np.sin(2 * np.pi * f2 * t)  # Otra señal sinusoidal

  # Convolución
  conv_result = np.convolve(sig1, sig2)  # Usamos 'same' para mantener la longitud

  # Graficar
  plt.figure(figsize=(12, 8))

  plt.subplot(3, 1, 1)
  plt.plot(t,sig1)
  plt.title("Señal 1")

  plt.subplot(3, 1, 2)
  plt.plot(t, sig2)
  plt.title("Señal 2")

  plt.subplot(3, 1, 3)
  plt.plot(tc,conv_result)

  plt.show()

f2_slider = widgets.FloatSlider(value=100, min=2, max=200, step=1, description="f2 (Hz):")

# Crear la interfaz interactiva
interactive_plot = interactive(plot_signals_and_convolution, f2=f2_slider)

# Mostrar el widget
interactive_plot
```

```python
def plot_signals_and_convolution_2(f2, margen, shift):
    fs = 5000  # Frecuencia de muestreo
    t = np.linspace(0, 1, fs)  # 1 segundo de duración
    tc = np.linspace(0, 1, fs*2-1)

    sig1 = 10 * np.sin(2 * np.pi * 5 * t) + 10 * np.sin(2 * np.pi * f2 * t)  # Señal combinada
    sig2 = 10 * np.sin(2 * np.pi * f2 * t)  # Otra señal sinusoidal

    # Definir el margen de tiempo para la convolución
    m_margen = int(margen * fs)
    start_idx = fs // 2 - m_margen // 2
    end_idx = fs // 2 + m_margen // 2


    # Aplicar el intervalo móvil a sig2
    shift_samples = int(shift * fs)
    sig2_limited = np.zeros_like(sig2)
    start_shift_idx = max(0, shift_samples - m_margen // 2)
    end_shift_idx = min(fs, shift_samples + m_margen // 2)
    sig2_limited[start_shift_idx:end_shift_idx] = sig2[start_shift_idx:end_shift_idx]

    # Convolución
    conv_result = np.convolve(sig1, sig2_limited, mode='full')
    tc_limited = np.linspace(0, 1, len(conv_result))

    # Graficar
    plt.figure(figsize=(12, 8))

    plt.subplot(3, 1, 1)
    plt.plot(t, sig1)
    plt.title("Señal 1 con margen aplicado")

    plt.subplot(3, 1, 2)
    plt.plot(t, sig2_limited)
    plt.title("Señal 2 con intervalo móvil aplicado")

    plt.subplot(3, 1, 3)
    plt.plot(tc_limited, conv_result)
    plt.title("Convolución en el margen de tiempo")

    plt.show()

f2_slider = widgets.FloatSlider(value=100, min=2, max=200, step=1, description="f2 (Hz):")
margin_slider = widgets.FloatSlider(value=0.2, min=0.01, max=1, step=0.01, description="Margen (s):")
shift_slider = widgets.FloatSlider(value=0.5, min=0, max=1, step=0.01, description="Desplazamiento (s):")

interactive_plot = interactive_output(plot_signals_and_convolution_2, {'f2': f2_slider, 'margen': margin_slider, 'shift': shift_slider})

display(VBox([f2_slider, margin_slider, shift_slider, interactive_plot]))

```
